```python
import pandas as pd
from financetracker import create_app, db
from financetracker.models import MainTypes, CurrencyExchanges, CurrencyUpdateStatus
import sqlalchemy as sa

app = create_app()
```

## MainTypes Table

```python
query = sa.select(MainTypes)

with app.app_context():
    result = pd.read_sql(sql=query, con=db.engine)

result
```

## CurrencyUpdateStatus Table

- This table will only have one Entry. After each exchange update, the existing row will be overwritten with current status data.
- Current DateTime is utc.

```python
query = sa.select(CurrencyUpdateStatus)

with app.app_context():
    result = pd.read_sql(sql=query, con=db.engine)

result
```
